```python
import xarray
import pathlib
import matplotlib.pyplot as plt
```

```python
rooftops_path = pathlib.Path('../../workflow/output/skopje-z19/rooftops')
```

```python
ds = xarray.open_dataset(str(rooftops_path / 'rooftops.nc'))
```

```python
ds.info()
```

```python
print(f'Total rooftop area: {ds.section_area.sum().item()} square meters.')
print(f'Total superstructures area: {ds.superst_area.sum().item()} square meters.')
```

```python
plt.title('Section area distribution')
plt.hist(ds.section_area, bins=50)
plt.show()
plt.title('Section azimuth distribution')
plt.hist(ds.section_azimuth, bins=50)
plt.show()
plt.title('Section tilt distribution')
plt.hist(ds.section_tilt, bins=50)
plt.show()
plt.title('Superst classes')
plt.hist(ds.superst_seg_class, bins=9)
plt.show()
plt.title('Section flat')
plt.hist(ds.section_flat, bins=3)
plt.show()
```
